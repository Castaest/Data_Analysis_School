```python
import csv
import sys
import codecs
```

```python
csv.field_size_limit(2**24)
```

```python
vacs = list(csv.DictReader(codecs.open('Vacancy.csv', 'r', 'utf_8_sig')))
```

```python
print(f'В файле {len(vacs)} вакансии')
```

```python
for k in vacs[0].keys(): print(k, end = ', ')
```

```python
# iterate through vacantions
cnt_a = 0
cnt_ap = 0
cnt_i = 0
cnt_m = 0
dates = []
for d in vacs:
    if (('исследов') in d['vacdescription'].lower() and ('аналитик данных' in d['vactitle'].lower() or ('data analyst' in d['vactitle'].lower()))):
        cnt_i += 1
        dates.append(d['vacdate'])
    if ('аналитик данных' in d['vactitle'].lower() or ('data analyst' in d['vactitle'].lower())):
        cnt_a += 1
        if 'python' in d['vacdescription'].lower():
            cnt_ap += 1
    if ('data scientist' in d['vactitle'].lower() or ('экономист' in d['vactitle'].lower())):        
        cnt_m += 1
        
print(f'В вакансиях найдены {cnt_a} вакансии аналитика данных, в т.ч. {cnt_ap} вакансии в описании которых упоминается Python\n')  
print(f'Мне могут понравится {cnt_i} вакансий, но самая свежая из них датирована {max(dates)}, т.е. она, видимо, не актуальна\n')
print(f'Всего я могу работать на позициях {cnt_m + cnt_a} вакансий')
```

```python
# Сколько вакансий с позициями на которых вы работаете? Проверка, а какие вообще вакансии есть в файле. 
from collections import Counter

Counter(vac['vactitle'] for vac in vacs).most_common()
```
